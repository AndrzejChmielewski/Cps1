# Polecenie 1

Krok 1: Zdefmiuj czestotliwosc sygnalu sinusoidalnego f = 5 Hz, czestotliwosc próbkowania fs = 1000 Hz oraz poczatek trwania sygnalu 3 = 0 i
koniec b = 2 s.

Krok 2: Utwórz wektor czasu o nazwie t, trwajace od a do b. 2 krokami czasowymi wynoszacymi 1/fs.

Krok 3: Utwórz sygnal sinusoidalny sig1 za pomoca funkcji sin:

s = sin(2 * pi * f *t)

Krok 4: Wyswietl wykres zdefiniowanego sygnalu. Nazwij wykres i osie.


```python
import numpy as np
import matplotlib.pyplot as plt

#definiujemy częstoliwosc sygnału sinusoidalnego
f = 2

#Definiowanie czesttliwosci próbkowania
fs = 1000 # Hz

# Definiujemy poczqtek i koniec sygnalu
a = 0
b = 2

print(f'f = {f}')
print(f'fs = {fs}')
print(f'a = {a}')
print(f'b = {b}')
```

```python
# Utwórz wektor czasu od 0 do 2 sekundy
t = np.arange(a, b, 1/fs)

print(f't = {t}')
```

```python
# Utwórz sygnal sinusoidalny

sig1 = np.sin(2 * np.pi * f * t)

print(f's = {sig1}')
```

```python
# Wyswietlanie
plt.plot(t, sig1)
plt.grid(alpha=0.2)
plt.title('sygnal sinusoidalny 2Hz')
plt.xlabel('czas [s]')
plt.ylabel('amplituda')
plt.show( )
```

Powszechnie jest dodawanie szumu do generowanych sygnałów. Szum sprawia, że sygnały są bardziej realistyczne i zapewnia, że techniki przetwarzania sygnału są odporne na zaszumione dane. Aby dodać szum, można utworzyć wektor liczb losowych o takim samym rozmiarze jak sygnał

Krok 1: Utwórz wektor liczb losowych o rozkładzie normalnym o nazwie "noise", który ma taki sam rozmiar jak "sig".

```python
# Rozkład normalny, średnia = 0, odchylenie standardowe = 1
noise = np.random.normal(0, 1, size=sig1.size)
```

```python
# Mnożenie szumu (noise) przez 0.1
noise *= 0.1
```

```python
# Zaszumiony sygnał
sigNoisy = sig1 + noise
```

```python
# Wyświetlanie
plt.plot(t, sigNoisy)
plt.grid(alpha=0.2)
plt.title('Sygnał sinusoidalny zaszumiony 2Hz')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.show()
```

# Polecenie 3

Krok 1: Utwórz sygnał prostokątny o nazwie sq (zdefiniowany w Poleceniu 1) o częstotliwości 2 Hz.

Krok 2: Utwórz wykres sq w funkcji czasu (t), nazwij wykres i oznacz osie.

```python
from scipy import signal

# Definicja częstotliwości sygnału
fsq = 2

# Utwórz sygnał prostokątny
sq = signal.square(2 * np.pi * fsq * t)
```

```python
# Wyswietlanie
plt.plot(t, sq)
plt.grid(alpha=0.2)
plt.title('Sygnał prostokątny')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.show()
```

# Polecenie 6

Zdefiniuj częstotliwość próbkowania fs = 100 Hz. Następnie stwórz sumę następujących sygnałów sinusoidalnych:

Sygnał o częstotliwości f1 = 1 Hz, amplitudzie amplituda1 = 1,
Sygnał o częstotliwości f2 = 3 Hz, amplitudzie amplituda2 = 1/3,
Sygnał o częstotliwości f3 = 5 Hz, amplitudzie amplituda3 = 1/5.

```python
# Zmiana częstotliwości próbkowania
fs = 100
t = np.arange(a, b, 1/fs)

# Sygnał 1 Hz
f1 = 1
amplituda1 = 1
sig_1 = np.sin(2 * np.pi * f1 * t) * amplituda1

# Sygnał 3 Hz
f2 = 3
amplituda2 = 1/3
sig_2 = np.sin(2 * np.pi * f2 * t) * amplituda2

# Sygnał 5 Hz
f3 = 5
amplituda3 = 1/5
sig_3 = np.sin(2 * np.pi * f3 * t) * amplituda3

# Sygnał sumacyjny
sum_sig = sig_1 + sig_2 + sig_3

# Wyświetlanie
plt.plot(t, sum_sig)
plt.grid(alpha=0.2)
plt.title('Suma sygnałów')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.show()
```

# Polecenie 7

Na podstawie rozwiązania polecenia 6 zsumuj więcej niż trzy sinusoidy o rosnącej częstotliwości (tylko nieparzyste częstotliwości sygnału) i malejącej amplitudzie. Użyj pętli for, aby wykonać to zadanie.

```python
# Parametry sinusoid
frequencies = [1, 3, 5, 7, 9]  # Częstotliwości
amplitudes = [5, 4, 3, 2, 1]   # Amplitudy
sum_signal = np.zeros(len(t))

for f, a in zip(frequencies, amplitudes):
    sig = np.sin(2 * np.pi * f * t) * a
    sum_signal += sig

# Wyświetlanie
plt.plot(t, sum_signal)
plt.grid(alpha=0.2)
plt.title('Suma sygnałów')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.show()
```

# Polecenie 8

Zdefiniuj częstotliwość próbkowania fs = 10 Hz. Następnie utwórz sumę sygnałów sinusoidalnych o częstotliwościach 1 Hz i 1,05 Hz. Co możesz zaobserwować?



```python
# Zmiana częstotliwości próbkowania
fs = 10
a = 0
b = 2
t = np.arange(a, b, 1/fs)

# Sygnał 1 Hz
f1 = 1
sig_1 = np.sin(2 * np.pi * f1 * t)

# Sygnał 1,05 Hz
f2 = 1.05
sig_2 = np.sin(2 * np.pi * f2 * t)

# Sygnał sumacyjny
sum_sig = sig_1 + sig_2

# Wyświetlanie
plt.plot(t, sum_sig)
plt.grid(alpha=0.2)
plt.title('Suma sygnałów')
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')
plt.show()
```
